```python
import os
import re
from collections import defaultdict

def parse_filename(filename):
    """è§£æ 523_0.001.csv â†’ (523, 0.001)"""
    match = re.match(r"(\d+)_(\d+\.\d+)\.csv", filename)
    if match:
        temp = int(match.group(1))
        strain = float(match.group(2))
        return temp, strain
    return None, None

def select_files(file_list):
    # è§£ææ–‡ä»¶ï¼ŒæŒ‰æ¸©åº¦ + åº”å˜é€Ÿç‡åˆ†ç±»
    data = defaultdict(dict)
    for f in file_list:
        temp, strain = parse_filename(f)
        if temp is not None:
            data[temp][strain] = f

    # å®šä¹‰é€‰æ‹©é¡ºåºï¼ˆå›ºå®šè§„åˆ™ï¼‰
    temp_order_round = [
        [523, 673],  # ç¬¬ 1 è½®
        [573],       # ç¬¬ 2 è½®
        [623],       # ç¬¬ 3 è½®
        [523],       # ç¬¬ 4 è½®ï¼ˆå‰©ä½™ï¼‰
        [573],       # ç¬¬ 5 è½®
        [623],       # ç¬¬ 6 è½®
        [673],       # ç¬¬ 7 è½®
    ]

    must_rates = {0.001, 1.000}  # å¿…é€‰ä¸¤ç§é€Ÿç‡
    selected = []

    # ä¾æ¬¡æ‰§è¡Œ 7 è½®æŒ‘é€‰
    for round_idx, temps in enumerate(temp_order_round, 1):
        for t in temps:
            if t not in data:
                continue
            rates = data[t]

            if round_idx <= 3:  # å‰ä¸‰è½®ï¼šåªé€‰ 0.001 & 1.000
                for r in must_rates:
                    if r in rates and rates[r] not in selected:
                        selected.append(rates[r])
            else:  # åå››è½®ï¼šé€‰å‰©ä½™é€Ÿç‡
                for r, fname in rates.items():
                    if r not in must_rates and fname not in selected:
                        selected.append(fname)
        print(f"ğŸŸ¢ å®Œæˆç¬¬ {round_idx} è½®é€‰æ‹©ï¼Œå½“å‰å·²é€‰ {len(selected)} ä¸ªæ–‡ä»¶ã€‚")
        print("   å½“å‰é€‰æ‹©é¡ºåºï¼š", selected)
    return selected


# ================== ç¤ºä¾‹è¿è¡Œ ==================
if __name__ == "__main__":
    # è¯»å–ç›®å½•ä¸‹æ‰€æœ‰ csv æ–‡ä»¶
    csv_files = [f for f in os.listdir(".") if f.endswith(".csv")]
    ordered = select_files(csv_files)

    print("\nâœ… é€‰æ‹©ç»“æœï¼ˆæŒ‰é¡ºåºï¼‰ï¼š")
    for f in ordered:
        print(f)
```
